# IQR

```python
q1 = np.percentile(netflix['Cost_Per_Month_Standard'], 25, interpolation = 'midpoint')
q2 = np.percentile(netflix['Cost_Per_Month_Standard'], 50, interpolation = 'midpoint')
q3 = np.percentile(netflix['Cost_Per_Month_Standard'], 75, interpolation = 'midpoint')
```

```python
IQR = q3 - q1
print('interquartile range is', IQR)
low_lim = q1 - 1.5 * IQR
up_lim = q3 + 1.5 * IQR
print('low_limit is', low_lim)
print('up_limit is', up_lim)
outlier = []

for x in netflix['Cost_Per_Month_Standard']:
    if ((x > up_lim) or (x < low_lim)):
        outlier.append(x)
print('outlier in the dataset is',  outlier)
```
